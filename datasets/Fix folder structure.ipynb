```python
for x in glob.glob("VinBigdata/dataset/dataset/dataset/*/*"):
    source_dir=x
    target_dir=x.replace('VinBigdata/dataset/dataset/dataset/','VinBigdata/dataset/')
    print(source_dir)
    print(target_dir)
    try:
        os.makedirs(target_dir)
        file_names = os.listdir(source_dir)
        for file_name in file_names:
            shutil.move(os.path.join(source_dir, file_name), target_dir)
    except:
        file_names = os.listdir(source_dir)
        for file_name in file_names:
            shutil.move(os.path.join(source_dir, file_name), target_dir)
```
